```python
# prompt: download dataset from https://huggingface.co/datasets/surrey-nlp/PLOD-CW-25

!pip install datasets

from datasets import load_dataset

dataset = load_dataset("surrey-nlp/PLOD-CW-25")

dataset
```

```python
# Print the first 10 validation samples
print_set = dataset['validation']
print_len = 10
print_all = False   # Set to True to print the whole text including 'O' tag

for index in range(print_len):
  print(index)
  for token, pos_tag, ner_tag in zip(print_set['tokens'][index], print_set['pos_tags'][index], print_set['ner_tags'][index]):
    if print_all or 'AC' in ner_tag or 'LF' in ner_tag:
      print(token, pos_tag, ner_tag)
  print()
```
